##**List Comprehension**

basic syantax:- [expression for item in iterable]



```python
num=[]
for i in range(10):
  num.append(i**2)
print(num)


square=[i**2 for i in range(10)]
print(square)
```

with conditional logic:-[expression for item in iterable if condition]


```python
lst=[]
for i in range(10):
  if i%2==0:
    lst.append(i)
print(lst)

##this is list comprehension
even_num=[i for i in range(10) if i%2==0]
print(even_num)
```

NESTED LIST COMPREHENSION:-[expression for item1 in itereable for item2 in itereable2]

```python
lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]
print(pair)
```

#** factorial wala sawal **

```python
def factorial(n):
  if(n==0 or n==1):
    return 1
  else :
    return n*factorial(n-1);

print(factorial(5))
```
